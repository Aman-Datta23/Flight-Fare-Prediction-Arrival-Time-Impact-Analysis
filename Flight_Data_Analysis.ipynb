```python
# Showing all the Airlines with their Number of Flights in Horizontal Bar Graph

#data['airline'].value_counts(ascending=True).plot.barh( color = ['lightgreen', 'lightblue'])

airline_counts = data['airline'].value_counts(ascending=True)

# Generate a list of distinct colors for each airline
colors = plt.cm.tab20.colors  # This provides 20 distinct colors
color_list = colors[:len(airline_counts)]  # Slice to match the number of airlines

# Plot horizontal bar chart
airline_counts.plot.barh(color=color_list)
plt.title("Airlines with frequencies")

plt.xlabel(" Number of flights")

plt.ylabel(" Airlines")

plt.show()
```

Bar Graphs representing the Departure Time & Arrival Time

```python
data.head ()
```

```python
# Showing the Departure Time for the flights

data['departure_time'].value_counts()
```

```python
# Showing the Arrival Time for the flights

data['arrival_time'].value_counts()
```

```python
# Get counts for departure and arrival time categories
departure_counts = data['departure_time'].value_counts()
arrival_counts = data['arrival_time'].value_counts()

# Get number of bars for each
num_departure = len(departure_counts)
num_arrival = len(arrival_counts)

# Generate different colors using a colormap
departure_colors = plt.cm.Set3(range(num_departure))  # You can use other colormaps too like Set2, tab10, etc.
arrival_colors = plt.cm.Pastel1(range(num_arrival))

# Plotting
plt.figure(figsize=(16, 4))

# Departure Time Plot
plt.subplot(1, 2, 1)
plt.bar(departure_counts.index, departure_counts.values, color=departure_colors)
plt.title("Departure Time")
plt.xlabel("D. Time")
plt.ylabel("D. Freq")

# Arrival Time Plot
plt.subplot(1, 2, 2)
plt.bar(arrival_counts.index, arrival_counts.values, color=arrival_colors)
plt.title("Arrival Time")
plt.xlabel("A. Time")
plt.ylabel("A. Freq")

plt.tight_layout()
plt.show()
```

```python
source_counts = data['source_city'].value_counts()
destination_counts = data['destination_city'].value_counts()

# Number of cities
num_source = len(source_counts)
num_dest = len(destination_counts)

# Generate different colors for each bar using colormaps
source_colors = plt.cm.Set2(range(num_source))  # or try Set3, Pastel1, etc.
dest_colors = plt.cm.Paired(range(num_dest))

# Plotting
plt.figure(figsize=(16, 4))

# Source City Bar Chart
plt.subplot(1, 2, 1)
plt.barh(source_counts.index, source_counts.values, color=source_colors)
plt.title("Source Cities with No. of flights")
plt.ylabel("Cities")
plt.xlabel("No. of flights")

# Destination City Bar Chart
plt.subplot(1, 2, 2)
plt.barh(destination_counts.index, destination_counts.values, color=dest_colors)
plt.title("Destination Cities with No. of flights")
plt.ylabel("Cities")
plt.xlabel("No. of flights")

plt.tight_layout()
plt.show()
```

Price varies with airlines ?

```python
# Grouping the airlines and checking their mean price

data.groupby('airline')['price'].mean()
```

```python
# Drawing a Categorical Plot showing the Mean Ticket Price for each Airline

# Optional: Clean the 'class' column to fix duplicates
data['class'] = data['class'].str.strip().str.capitalize()

# Get unique number of airlines for palette size
unique_airlines = data['airline'].nunique()

# Create a distinct color palette with enough colors
palette = sns.color_palette("hsv", unique_airlines)

# Plot using 'col' instead of 'hue' if you only want to vary color by airline
sns.catplot(
    x='airline',
    y='price',
    kind='bar',
    data=data,
    palette=palette,
    height=5,
    aspect=2
)

plt.title("Mean Ticket Price for Each Airline")
plt.xlabel("Airline")
plt.ylabel("Mean Price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Does ticket price change based on the departure time and arrival time?

```python
# Checking the Mean Ticket Price based on the Departure Times

data.groupby('departure_time')['price'].mean()
```

```python
# Checking the Mean Ticket Price based on the Arrival Times

data.groupby('arrival_time')['price'].mean()
```

```python
# Get the number of unique departure times
unique_departures = data['departure_time'].nunique()

# Generate a distinct color for each departure time
custom_palette = sns.color_palette("Set2", unique_departures)  # You can also use Set3, Pastel1, tab10, etc.

# Plot with palette by mapping to 'departure_time'
sns.catplot(
    x='departure_time',
    y='price',
    kind='bar',
    data=data,
    palette=dict(zip(data['departure_time'].unique(), custom_palette))
)

plt.title("Average Ticket Price by Departure Time")
plt.xlabel("Departure Time")
plt.ylabel("Average Price")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

```python
# sns.catplot( x = 'departure_time', y = 'price', kind = 'bar', data = data )

# plt.show()
```

```python
# Get number of unique arrival time categories
unique_arrival_times = data['arrival_time'].nunique()

# Generate a unique color for each category
custom_palette = sns.color_palette("Set3", unique_arrival_times)  # Try Set3, Set2, tab10, etc.

# Map each arrival_time to a color
palette_dict = dict(zip(data['arrival_time'].unique(), custom_palette))

# Create the plot
sns.catplot(
    x='arrival_time',
    y='price',
    kind='bar',
    data=data,
    palette=palette_dict
)

plt.title("Average Ticket Price by Arrival Time")
plt.xlabel("Arrival Time")
plt.ylabel("Average Price")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```
